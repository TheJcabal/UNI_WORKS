### 1.1 Crear el DataFrame utilizando Pandas

`pd.DataFrame()` convierte el diccionario de datos en una estructura tabular de filas y
columnas. Cada llave del diccionario se convierte en una columna y cada valor en los
registros de esa columna.

## Punto 2 – Operaciones Matemáticas entre Columnas

En esta sección se crean nuevas columnas derivadas de operaciones aritméticas entre columnas
existentes. Pandas permite operar columnas directamente como si fueran vectores, aplicando
la operación elemento a elemento sobre todos los registros del DataFrame.

### 2.1 Nueva columna: Total_Venta = Precio x Cantidad

Total_Venta representa el ingreso bruto generado por cada producto: el precio unitario
multiplicado por las unidades disponibles. Esta columna es la base para todos los cálculos
financieros posteriores del taller.

```python
# 2.1 Crear una nueva columna llamada Total_Venta = Precio × Cantidad
df["Total_Venta"] = df["Precio"] * df["Cantidad"]
print(
    f"Total_Venta:\n{df['Total_Venta']}"
)  # muestra el DataFrame con la nueva columna Total_Venta
```

### 2.2 Nueva columna: Costo_Total = Total_Venta + Costo_Envio

Costo_Total agrega al ingreso bruto el costo logístico asociado al envío del producto.
Permite tener una visión más completa del valor real que representa cada producto,
considerando no solo el precio sino también los costos operativos de distribución.

```python
# 2.2 crear una nueva columna llamada Costo_Total = Total_Venta + Costo_Envio.
df["Costo_Total"] = df["Total_Venta"] + df["Costo_Envio"]
print(
    f"Costo_Total:\n{df['Costo_Total']}"
)  # muestra el DataFrame con la nueva columna Costo_Total
```

### 2.3 Promedio del Total_Venta

El promedio o media aritmética de Total_Venta indica el ingreso bruto típico por producto.
`mean()` suma todos los valores y los divide entre el número de registros. Sirve como
referencia para identificar qué productos están por encima o por debajo del rendimiento
promedio del catálogo.

```python
# 2.3 Calcular el promedio del Total_Venta.
promedio_total_venta = df["Total_Venta"].mean()  # calcula el promedio de Total_Venta
print(
    f"Promedio de Total_Venta:\n{promedio_total_venta}"
)  # muestra el promedio de Total_Venta
```

### 2.4 Producto con mayor Total_Venta

Se filtra el DataFrame comparando cada valor de Total_Venta contra el máximo de la columna,
obtenido con `max()`. El resultado identifica cuál producto genera el mayor ingreso bruto,
información clave para decisiones de inventario y estrategia comercial.

```python
# 2.4 Identificar el producto con mayor Total_Venta.
producto_mayor_venta = df[
    df["Total_Venta"] == df["Total_Venta"].max()
]  # identifica el producto con mayor Total_Venta
print(
    f"Producto con mayor Total_Venta:\n{producto_mayor_venta}"
)  # muestra el producto con mayor Total_Venta
```

### 2.5 Producto con menor Total_Venta

De forma análoga al punto anterior, se filtra usando `min()` para identificar el producto
con el ingreso bruto más bajo. Puede indicar artículos con baja rotación, precio muy
reducido o stock mínimo que requieren revisión en la estrategia de ventas.

```python
# 2.5 Identificar el producto con menor Total_Venta
producto_menor_venta = df[
    df["Total_Venta"] == df["Total_Venta"].min()
]  # identifica el producto con menor Total_Venta
print(
    f"Producto con menor Total_Venta:\n{producto_menor_venta}"
)  # muestra el producto con menor Total_Venta
```

## Punto 3 – Porcentajes y Análisis

En esta sección se crean columnas que expresan valores del Total_Venta como porcentajes,
lo que permite analizar la participación y el impacto financiero de cada producto dentro
del conjunto de datos.

### 3.1 Nueva columna: IVA = 19% del Total_Venta

En Colombia, el Impuesto al Valor Agregado (IVA) para productos tecnológicos es del 19%.
Esta columna calcula el valor del IVA que corresponde a cada producto en función de su
Total_Venta, multiplicando por 0.19.

```python
# 3.1 Crear una columna llamada IVA que represente el 19% del Total_Venta
df["IVA"] = df["Total_Venta"] * 0.19  # crea la columna IVA como el 19% de Total_Venta
print(f"IVA:\n{df['IVA']}")  # muestra el DataFrame con la nueva columna IVA
```

### 3.2 Nueva columna: Ganancia_Estimada = 25% del Total_Venta

La ganancia estimada representa el margen de utilidad proyectado para cada producto,
equivalente al 25% del ingreso bruto. Este porcentaje es un supuesto del negocio para
estimar la rentabilidad antes de descontar costos fijos y variables adicionales.

```python
# 3.2 Crear una columna llamada Ganancia_Estimada correspondiente al 25% del Total_Venta
df["Ganancia_Estimada"] = (
    df["Total_Venta"] * 0.25
)  # crea la columna Ganancia_Estimada como el 25% de Total_Venta
print(
    f"Ganancia_Estimada:\n{df['Ganancia_Estimada']}"
)  # muestra el DataFrame con la nueva columna Ganancia_Estimada
```

### 3.3 Nueva columna: Perdida_Estimada = 0.05% del Total_Venta

La pérdida estimada representa una fracción muy pequeña del ingreso bruto (0.05%),
que modela posibles mermas, devoluciones o deterioro de productos. Aunque el porcentaje
es mínimo por producto, en catálogos grandes su suma acumulada puede ser significativa.

```python
# 3.3 Crear una columna llamada Perdida_Estimada correspondiente al 0.05% del Total_Venta.
df["Perdida_Estimada"] = (
    df["Total_Venta"] * 0.0005
)  # crea la columna Perdida_Estimada como el 0.05% de Total_Venta
print(
    f"Perdida_Estimada:\n{df['Perdida_Estimada']}"
)  # muestra el DataFrame con la nueva columna Perdida_Estimada
```

### 3.4 Porcentaje de cada producto respecto al total general de ventas

Se calcula el total general sumando todos los valores de Total_Venta con `sum()`.
Luego se divide el Total_Venta de cada producto entre ese total y se multiplica por 100.
El resultado indica qué fracción del ingreso total representa cada artículo, permitiendo
identificar los productos más relevantes para el negocio.

```python
# 3.4 Calcular qué porcentaje representa cada producto respecto al total general de ventas.
total_general_ventas = df["Total_Venta"].sum()  # calcula el total general de ventas
df["Porcentaje_Ventas"] = (
    df["Total_Venta"] / total_general_ventas * 100
)  # crea la columna Porcentaje_Ventas como el porcentaje de cada producto respecto al total general de ventas
print(
    f"Porcentaje de Ventas:\n{df['Porcentaje_Ventas']}"
)  # muestra el DataFrame con la nueva columna Porcentaje_Ventas
```

### 3.5 Productos con porcentaje de ventas superior al 15%

Se filtra el DataFrame para mostrar únicamente los productos cuya participación en ventas
supera el 15% del total general. Un umbral del 15% es significativo en un catálogo de 60
productos, donde una distribución uniforme implicaría aproximadamente 1.67% por producto.
Superar ese umbral indica productos dominantes en la estructura de ingresos.

```python
# 3.5 Mostrar únicamente los productos cuyo porcentaje de ventas sea superior al 15%
productos_superiores_15 = df[
    df["Porcentaje_Ventas"] > 15
]  # filtra los productos cuyo porcentaje de ventas es superior al 15%
print(
    f"Productos con porcentaje de ventas superior al 15%:\n{productos_superiores_15}"
)  # muestra los productos cuyo porcentaje de ventas es superior al 15%
```

## Punto 4 – Estadística Descriptiva

La estadística descriptiva permite caracterizar el comportamiento de los datos mediante
medidas de tendencia central y dispersión. En esta sección se analizan las columnas
Precio, Cantidad y Total_Venta para entender su distribución e identificar posibles
valores extremos o comportamientos inusuales en el dataset.

### 4.1 Media, Mediana y Moda del Precio

Se calculan las tres medidas de tendencia central sobre la columna Precio.
`mean()` calcula el promedio aritmético, `median()` el valor central de la distribución
ordenada, y `mode()` el valor que aparece con mayor frecuencia. Comparar las tres permite
detectar si la distribución es simétrica o si está sesgada por valores extremos:
cuando la media es mayor que la mediana, existe sesgo hacia precios altos.

```python
# 4.1 Calcular la media, mediana y moda de la columna Precio
media_precio = df["Precio"].mean()  # calcula la media de la columna Precio
mediana_precio = df["Precio"].median()  # calcula la mediana de la columna Precio
moda_precio = df["Precio"].mode()[0]  # calcula la moda de la columna Precio
# imprimir los resultados de la media, mediana y moda del Precio
print(f"Media del Precio: {media_precio}")  # muestra la media del Precio
print(f"Mediana del Precio: {mediana_precio}")  # muestra la mediana del Precio
print(f"Moda del Precio: {moda_precio}")  # muestra la moda del Precio
```

### 4.2 Desviación estándar de la Cantidad

`std()` calcula la desviación estándar, que mide qué tan alejados están los valores
de la Cantidad respecto a su promedio. Un valor alto indica alta variabilidad en las
unidades entre productos, lo que puede reflejar diferencias de demanda, políticas de
stock o tipos de producto muy distintos dentro del catálogo.

```python
# 4.2 Calcular la desviación estándar de la columna Cantidad.
desviacion_estandar_cantidad = df[
    "Cantidad"
].std()  # calcula la desviación estándar de la columna Cantidad
print(
    f"Desviación Estándar de la Cantidad: {desviacion_estandar_cantidad}"
)  # muestra la desviación estándar de la Cantidad
```

### 4.3 Valor máximo y mínimo del Total_Venta

`max()` y `min()` retornan los valores extremos de la distribución de ingresos brutos.
La diferencia entre ambos (rango) indica la amplitud de la variación en ventas.
Un rango muy amplio señala que el catálogo incluye productos con perfiles de venta muy
diferentes, lo que puede requerir estrategias comerciales diferenciadas.

```python
# 4.3 Calcular el valor máximo y mínimo de Total_Venta
valor_maximo_total_venta = df[
    "Total_Venta"
].max()  # calcula el valor máximo de Total_Venta
valor_minimo_total_venta = df[
    "Total_Venta"
].min()  # calcula el valor mínimo de Total_Venta
print(
    f"Valor Máximo de Total_Venta: {valor_maximo_total_venta}"
)  # muestra el valor máximo de Total_Venta
print(
    f"Valor Mínimo de Total_Venta: {valor_minimo_total_venta}\n"
)  # muestra el valor mínimo de Total_Venta
```

### 4.4 Interpretación de las medidas estadísticas

**Media del Precio:** valor promedio de todos los precios del catálogo. Se ve desplazada
por productos de precio muy alto o muy bajo, por lo que debe interpretarse junto con
la mediana para evaluar si es representativa.

**Mediana del Precio:** valor central de la distribución de precios. Al no verse afectada
por valores atípicos, representa mejor el precio "típico" del catálogo que la media.

**Moda del Precio:** precio que aparece con mayor frecuencia. En este dataset, donde la
mayoría de precios son únicos, la moda tiene menor relevancia estadística pero puede
indicar rangos de precio estandarizados.

**Desviación estándar de la Cantidad:** mide la dispersión de las unidades respecto al
promedio. Un valor elevado indica que las cantidades varían considerablemente entre
productos, señalando diferencias importantes en volumen de inventario.

**Valor máximo de Total_Venta:** identifica el producto que genera el mayor ingreso bruto,
útil para priorizar atención en la gestión comercial y de inventario.

**Valor mínimo de Total_Venta:** identifica el producto de menor contribución al ingreso
total. Puede indicar candidatos a revisión de precio, cantidad o incluso discontinuación.

## Punto 5 – Filtrado

El filtrado permite extraer subconjuntos del DataFrame que cumplen condiciones específicas.
Pandas usa índices booleanos: se evalúa una condición sobre una columna, lo que genera
una serie de valores True/False, y el DataFrame retorna solo las filas donde la condición
es True.

### 5.1 Productos de la categoría Tecnología

Se filtra el DataFrame usando la condición `df['Categoria'] == 'Tecnologia'`. El resultado
muestra únicamente los productos clasificados en esa categoría, permitiendo analizar de
forma aislada el segmento tecnológico del catálogo, que representa la mayor parte del
inventario en este dataset.

```python
# 5.1 Mostrar únicamente los productos de la categoría Tecnología.
productos_tecnologia = df[
    df["Categoria"] == "Tecnologia"
]  # filtra los productos de la categoría Tecnología
print(
    f"Productos de la categoría Tecnología:\n{productos_tecnologia}"
)  # muestra los productos de la categoría Tecnología
```

### 5.2 Ordenar productos de mayor a menor según Total_Venta

`sort_values('Total_Venta', ascending=False)` reordena todas las filas del DataFrame de
mayor a menor ingreso bruto. El parámetro `ascending=False` indica orden descendente.
Este ordenamiento permite identificar rápidamente el ranking de productos por desempeño
comercial y priorizar decisiones de reabastecimiento o promoción.

```python
# 5.2 Ordenar los productos de mayor a menor según Total_Venta.
productos_ordenados = df.sort_values("Total_Venta", ascending=False)
# ordena los productos de mayor a menor según Total_Venta
# ascending=False indica que el orden es de mayor a menor
print(
    f"Productos ordenados de mayor a menor según Total_Venta:\n{productos_ordenados}"
)  # muestra los productos ordenados de mayor a menor según Total_Venta
```

## Punto Adicional – Exportación a Excel (.xlsx)

un archivo compatible con Microsoft Excel usando la librería `openpyxl` como motor de
escritura. El parámetro `index=False` evita que el índice numérico del DataFrame se
incluya como columna en el archivo exportado. En Google Colab, el archivo generado se
puede descargar usando `google.colab.files.download()`.
