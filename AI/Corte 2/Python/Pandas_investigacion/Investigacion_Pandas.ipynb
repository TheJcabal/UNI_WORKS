## 1. ¿Qué es Pandas?

**Pandas** es una librería de código abierto para el lenguaje de programación Python, diseñada para la **manipulación y análisis de datos**. Su nombre proviene de *"Panel Data"*, un término usado en econometría para conjuntos de datos multidimensionales.

Fue creada por **Wes McKinney** en 2008 y actualmente es mantenida por una amplia comunidad de desarrolladores. Es una de las herramientas más utilizadas en el mundo del **Data Science**, **Machine Learning** e **Inteligencia Artificial**.

Pandas se basa en **NumPy** y proporciona estructuras de datos de alto rendimiento y flexibles, junto con herramientas de análisis de datos fáciles de usar.

```
Python → NumPy (arrays numéricos) → Pandas (tablas y series de datos)
```

## 2. ¿Para qué sirve Pandas?

Pandas sirve para realizar múltiples tareas relacionadas con el manejo de datos:

| Tarea | Descripción |
|---|---|
| **Cargar datos** | Leer archivos CSV, Excel, JSON, SQL, HTML, etc. |
| **Limpiar datos** | Manejar valores nulos, duplicados, tipos incorrectos |
| **Transformar datos** | Filtrar, agrupar, ordenar, pivotar |
| **Analizar datos** | Estadísticas descriptivas, correlaciones |
| **Visualizar datos** | Integración con Matplotlib y Seaborn |
| **Exportar datos** | Guardar en CSV, Excel, JSON, bases de datos |

##3. Objetivo Principal de la Librería

El objetivo principal de Pandas es proporcionar **estructuras de datos rápidas, flexibles y expresivas** diseñadas para trabajar con datos relacionales o etiquetados de forma intuitiva y eficiente.

Busca ser el **bloque fundamental de alto nivel** para realizar análisis de datos prácticos y reales en Python, siendo accesible tanto para principiantes como para expertos en ciencia de datos.

Sus dos estructuras de datos principales son:

🔹 **Series:** Array unidimensional etiquetado (como una columna de Excel)

🔹 **DataFrame:** Estructura bidimensional etiquetada (como una tabla completa de Excel)

## 4. Principales Funciones y Características

### Estructuras de Datos
- `pd.Series()` → Crea una serie unidimensional
- `pd.DataFrame()` → Crea una tabla de datos

### Lectura y Escritura de Datos
- `pd.read_csv()` → Lee archivos CSV
- `pd.read_excel()` → Lee archivos Excel
- `df.to_csv()` → Exporta a CSV

### Exploración de Datos
- `df.head()` / `df.tail()` → Primeras / últimas filas
- `df.info()` → Información general del DataFrame
- `df.describe()` → Estadísticas descriptivas
- `df.shape` → Dimensiones (filas, columnas)

### Limpieza de Datos
- `df.dropna()` → Elimina filas con valores nulos
- `df.fillna()` → Rellena valores nulos
- `df.drop_duplicates()` → Elimina filas duplicadas

### Transformación
- `df.groupby()` → Agrupa datos
- `df.merge()` → Une dos DataFrames
- `df.sort_values()` → Ordena por columna
- `df.apply()` → Aplica una función a columnas/filas

### Estadísticas
- `df.mean()`, `df.sum()`, `df.max()`, `df.min()` → Operaciones estadísticas básicas
- `df.corr()` → Correlación entre columnas

##5. Importancia de Pandas en IA y Análisis de Datos

Pandas es **fundamental** en el flujo de trabajo de Inteligencia Artificial y Machine Learning porque:

1. **Preprocesamiento de datos:** Antes de entrenar cualquier modelo de IA, los datos deben ser limpiados y estructurados. Pandas es la herramienta ideal para esto.

2. **Integración con librerías de ML:** Pandas funciona perfectamente con Scikit-learn, TensorFlow, PyTorch, etc.

3. **Exploración de datos (EDA):** Permite entender la distribución, calidad y características de los datos antes del modelado.

4. **Eficiencia:** Procesa millones de registros de forma rápida gracias a su base en NumPy.

5. **Estandarización:** Es el formato estándar que utilizan las empresas y la industria para compartir y trabajar con datos.

## 🔹 Ejercicio 1: Crear un DataFrame desde cero

Se creará un DataFrame que representa información de estudiantes de una clase.

````python
# -------------------------------------------------------
# Ejercicio 1: Crear el DataFrame
# -------------------------------------------------------
# importar libreria  pandas
import pandas as pd

# crear un dataframe con pandas
data = {  # DICCIONARIO CON LOS DATOS
    "ANIMAL": ["Perro", "Gato", "Pájaro", "Pez"],
    "EDAD": [5, 3, 2, 1], # años
    "PESO": [20, 5, 0.5, 0.1], # kilo
}
df = pd.DataFrame(data)  # crear un dataframe a partir de un diccionario

# MOSTRAR EL DATAFRAME
print(df)
````

#Explicacion del punto 1


  1) Importa la librería Pandas: import pandas as pd carga la librería Pandas en el entorno, lo que permite usar sus funciones bajo el alias pd.

    
  2) Crea un diccionario de datos: Se define un diccionario llamado data. Este diccionario contiene tres claves ("ANIMAL", "EDAD", "PESO"), donde cada clave está asociada a una lista de valores. Estas listas representan las columnas de los datos que se van a crear.

  3)  Crea un DataFrame: df = pd.DataFrame(data) convierte el diccionario data en un objeto DataFrame de Pandas. Un DataFrame es una estructura de datos bidimensional con etiquetas, similar a una tabla de una hoja de cálculo, donde las claves del diccionario se convierten en los nombres de las columnas y las listas en las filas de datos.

  4) Imprime el DataFrame: print(df) muestra el contenido del DataFrame recién creado en la salida, lo que permite visualizar la tabla con los datos de animales, edad y peso.

## 🔹 Ejercicio 2: Operaciones entre Columnas

Se realizarán cálculos entre columnas: promedio de notas, aprobación y clasificación.

````python
# -------------------------------------------------------
# Ejercicio 2: Operaciones entre columnas
# -------------------------------------------------------

# Calcular la media de la columna 'EDAD'
media_edad = df["EDAD"].mean()
print(f"Media de edad: {media_edad}")

# SUMAR LOS PESOS DE LOS ANIMALES
suma_peso = df["PESO"].sum()
print(f"Suma de peso: {suma_peso}")
````

## EXPLICACIÓN PUNTO 2
  
  1) Calcula la media de la edad: Utiliza df["EDAD"].mean() para obtener el promedio de todos los valores en la columna 'EDAD' del DataFrame df y lo guarda en la variable media_edad.
    
  2) Imprime la media de la edad: Muestra por pantalla el valor calculado de la media de edad.
  
  3) Calcula la suma de los pesos: Emplea df["PESO"].sum() para obtener la suma total de todos los valores en la columna 'PESO' del DataFrame df y lo almacena en la variable suma_peso.

  4) Imprime la suma de los pesos: Muestra por pantalla el valor calculado de la suma de los pesos.
